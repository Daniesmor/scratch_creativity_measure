```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de ejemplo
x = [0, 0, 1, 2, 3, 4, 2, 3, 0]  # Coordenadas en el eje x
y = [0, 1, 0, 0, 0, 0, 1, 1, 2]  # Coordenadas en el eje y
labels = ['Initialize maximo to -', 
          'Initialize contador_maximos to 0',
          'For each number in the list', 
          'If the number is greater than maximo', 
          'Set maximo to the number', 
          'Reset contador_maximos to 1', 
          'If the number is equal to maximo', 
          'Increment contador_maximos by 1', 
          'Return maximo and contador_maximos'
         ]

plt.figure(figsize=(20, 5))
plt.xticks([i * 0.5 for i in range(int(max(x) / 0.5) + 2)])
plt.yticks([i * 0.5 for i in range(int(max(y) / 0.5) + 2)])

# Crear el grfico de puntos
plt.scatter(x, y, color='red', marker='o')

# Aadir etiquetas a cada punto
for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Aadir etiquetas y ttulo
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Algorithm 1 (Inefficient)')
plt.grid(True)

# Mostrar el grfico
plt.show()
```

```python
# Datos de ejemplo
x = [0, 1, 2, 3, 4, 5, 3, 4, 1]  # Coordenadas en el eje x
y = [0, 0, 0, 0, 0, 0, 1, 1, 1]  # Coordenadas en el eje y
labels = ['Initialize maximo to -',
          'Initialize contador_maximos to 0',
          'For each number in the list',
          'If the number is greater than maximo',
          'Set maximo to the number',
          'Reset contador_maximos to 1',
          'If the number is equal to maximo',
          'Increment contador_maximos by 1',
          'Return maximo and contador_maximos'
         ]

plt.figure(figsize=(20, 5))
plt.xticks([i * 0.5 for i in range(int(max(x) / 0.5) + 2)])

# Crear el grfico de puntos
plt.scatter(x, y, color='blue', marker='o')

# Aadir etiquetas a cada punto
for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Aadir etiquetas y ttulo
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Algorithm 2 (Efficient)')
plt.grid(True)

# Mostrar el grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas de los puntos en el grfico ineficiente
ineficiente = np.array([
    [0, 0],  # Inicializar maximo a -
    [0, 1],  # Inicializar contador_maximos a 0
    [1, 0],  # Para cada nmero en la lista
    [2, 0],  # Si el nmero es mayor que maximo
    [3, 0],  # Asignar maximo al nmero
    [4, 0],  # Reiniciar contador_maximos a 1
    [2, 1],  # Si el nmero es igual a maximo
    [3, 1],  # Incrementar contador_maximos en 1
    [0, 2]   # Retornar maximo y contador_maximos
])

# Coordenadas de los puntos en el grfico eficiente
eficiente = np.array([
    [0, 0],  # Inicializar maximo a -
    [1, 0],  # Inicializar contador_maximos a 0
    [2, 0],  # Para cada nmero en la lista
    [3, 0],  # Si el nmero es mayor que maximo
    [4, 0],  # Asignar maximo al nmero
    [5, 0],  # Reiniciar contador_maximos a 1
    [3, 1],  # Si el nmero es igual a maximo
    [4, 1],  # Incrementar contador_maximos en 1
    [1, 1]   # Retornar maximo y contador_maximos
])

# Calcular los vectores de cambio
vectores_cambio = eficiente - ineficiente

# Calcular la magnitud y el ngulo de los vectores de cambio
magnitudes = np.linalg.norm(vectores_cambio, axis=1)
ngulos = np.arctan2(vectores_cambio[:, 1], vectores_cambio[:, 0]) * 180 / np.pi  # En grados

# Imprimir resultados
for i, (vector, magnitud, ngulo) in enumerate(zip(vectores_cambio, magnitudes, ngulos)):
    print(f"Punto {i}:")
    print(f"  Vector de Cambio: {vector}")
    print(f"  Magnitud: {magnitud:.2f}")
    print(f"  ngulo: {ngulo:.2f} grados")

# Visualizacin de los vectores de cambio
plt.figure(figsize=(10, 6))

plt.quiver(ineficiente[:, 0], ineficiente[:, 1], vectores_cambio[:, 0], vectores_cambio[:, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Change vector')
plt.scatter(ineficiente[:, 0], ineficiente[:, 1], color='red', label='Algorithm 1', zorder=5)
plt.scatter(eficiente[:, 0], eficiente[:, 1], color='blue', label='Algorithm 2', zorder=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Point changes between Algorithm 1 and Algorithm 2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Coordenadas de los puntos en el grfico ineficiente
ineficiente = np.array([
    [0, 0],  # Inicializar suma_total
    [0, 1],  # Inicializar contador_pares
    [2, 1],  # Para cada nmero en la lista
    [3, 1],  # Si el nmero es par
    [4, 1],  # Incrementar contador_pares en 1
    [3, -1], # Agregar nmero a suma_total
    [1, -2]  # Retornar suma_total y contador_pares
])

# Coordenadas de los puntos en el grfico eficiente
eficiente = np.array([
    [0, 0],  # Inicializar suma_total
    [1, 0],  # Inicializar contador_pares
    [2, 1],  # Para cada nmero en la lista
    [3, 1],  # Agregar nmero a suma_total
    [4, 1],  # Si el nmero es par
    [5, 1],  # Incrementar contador_pares en 1
    [1, -1]  # Retornar suma_total y contador_pares
])

# Calcular los vectores de cambio
vectores_cambio = eficiente - ineficiente

# Calcular la magnitud y el ngulo de los vectores de cambio
magnitudes = np.linalg.norm(vectores_cambio, axis=1)
ngulos = np.arctan2(vectores_cambio[:, 1], vectores_cambio[:, 0]) * 180 / np.pi  # En grados

# Imprimir resultados
for i, (vector, magnitud, ngulo) in enumerate(zip(vectores_cambio, magnitudes, ngulos)):
    print(f"Punto {i}:")
    print(f"  Vector de Cambio: {vector}")
    print(f"  Magnitud: {magnitud:.2f}")
    print(f"  ngulo: {ngulo:.2f} grados")

# Visualizacin de los vectores de cambio
plt.figure(figsize=(10, 6))

plt.quiver(ineficiente[:, 0], ineficiente[:, 1], vectores_cambio[:, 0], vectores_cambio[:, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vectores de Cambio')
plt.scatter(ineficiente[:, 0], ineficiente[:, 1], color='blue', label='Ineficiente', zorder=5)
plt.scatter(eficiente[:, 0], eficiente[:, 1], color='green', label='Eficiente', zorder=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Cambio de Puntos entre Ineficiente y Eficiente')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
